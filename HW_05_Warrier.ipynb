```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/Users/swarrier/Documents/CS301/student_class.csv')
```

## Problem 1

```python
df2 = df.groupby(['Majors'])
df3 = df2['Midterm'].median()
df4 = df2['Final'].median()
df5 = df2['Final Average'].median()

plt.rcParams["figure.figsize"] = (13, 8) # (w, h)
N = 5
x = np.arange(len(df['Majors'].unique()))
ind = np.arange(N)  
width = 0.25   
fig = plt.figure()
ax = fig.add_subplot(111)
rects1 = ax.bar(ind-width, df3, width, color='red', label = 'Midterm')
rects2 = ax.bar(ind, df4, width, color='green', label = 'Final')
rects3 = ax.bar(ind+width, df5, width, color='blue', label = 'Final Average')

ax.set_xticks(x + width / 2)
ax.set_xticklabels(df['Majors'].unique())

plt.ylabel('Median Score')
plt.xlabel('Major')
plt.title('Median Midterm, Final, and Final Average Scores')
plt.legend()
```

## Problem 2

```python
plt.rcParams["figure.figsize"] = (15, 7) # (w, h)

plt.hist(df['Numeric Bin'], bins=6)

plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.title("Number of Students in Each Grade Bin")
```

## Problem 3 

```python
plt.rcParams["figure.figsize"] = (15, 7) # (w, h)

plt.plot(df['Midterm'], df['Final Average'], 'o', label='Midterm')
plt.plot(df['Final'], df['Final Average'], 'o', label='Final')
plt.gcf().set_size_inches(12, 6)
plt.ylabel('Final Average')
plt.xlabel('Exam Score')
plt.title('Exams vs Final Average')
plt.tight_layout()
plt.legend()
plt.show()
```
